## Task 1 — Scatter: life expectancy change over time

**What to build:** A scatter showing **GDP per capita vs life expectancy** for **two years** (2000 and 2007) to show how both moved — use **colour for year** (just 2 colours), **one continent only**.

Choose any continent except Africa (that was the example). Highlight the change direction.

> 💡 Filter: `df.loc[df['continent'] == 'YOUR_CHOICE']` then filter years


```python
# Task 1 — Scatter Chart

import pandas as pd
import plotly.express as px

# Load dataset
df = pd.read_csv("/content/gapminder.csv")

# Create scatter plot
fig = px.scatter(
    df,
    x="GDP_per_capita",
    y="Life_expectancy",
    color="Continent",                     # one categorical colour variable
    hover_name="Country",
    log_x=True,                            # required log scale
    title="Countries With Higher GDP Tend to Have Longer Life Expectancy",
    color_discrete_sequence=px.colors.qualitative.Safe
)

# Improve layout
fig.update_layout(
    template="simple_white",
    xaxis_title="GDP per Capita (log scale)",
    yaxis_title="Life Expectancy",
    legend_title="Continent"
)

# Show chart
fig.show()
```

## Task 2 — Bubble chart: tell a story

**What to build:** A bubble chart (full 2007 dataset, all countries) where:
- x = GDP per capita (log scale)
- y = life expectancy
- size = population
- colour = ONE continent highlighted (your choice), all others grey
- At least one annotation explaining the highlighted group's story

> This is the grey-and-highlight technique applied to a bubble chart.


```python
# Task 2 — Bubble Chart: Tell a Story

import pandas as pd
import plotly.express as px

# Load dataset
df = pd.read_csv("/content/gapminder.csv")

# Highlight one continent (example: Asia)
df["Highlight"] = df["Continent"].apply(
    lambda x: "Asia" if x == "Asia" else "Other"
)

# Create bubble chart
fig = px.scatter(
    df,
    x="GDP_per_capita",
    y="Life_expectancy",
    size="Population",
    color="Highlight",
    hover_name="Country",
    log_x=True,                           # required
    size_max=60,                          # required
    title="Asian Countries Show Wide Differences in Wealth and Life Expectancy",
    color_discrete_map={
        "Asia": "#1f77b4",
        "Other": "lightgrey"
    }
)

# Add annotation
fig.add_annotation(
    x=5000,
    y=73,
    text="Asia includes both rapidly developing<br>and lower-income countries",
    showarrow=True,
    arrowhead=2
)

# Improve layout
fig.update_layout(
    template="simple_white",
    xaxis_title="GDP per Capita (log scale)",
    yaxis_title="Life Expectancy",
    legend_title=""
)

# Show chart
fig.show()
```
